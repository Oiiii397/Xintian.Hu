```python
df.groupby("Main_Note")["Rating"].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

note_rating = df.groupby("Main_Note")["Rating"].mean().sort_values(ascending=False)

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12,6))
note_rating.plot(kind="bar",color="coral")
plt.title("不同香调的香水评分排名",fontsize=14)
plt.xlabel("香调",fontsize=12)
plt.ylabel("平均评分",fontsize=12)
plt.xticks(rotation=45,ha="right")
plt.ylim(4,5)
plt.grid(axis="y",alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(10,6))
plt.scatter(df["Price_USD"],df["Rating"],alpha=0.7,colour="steelblue",s=100)
plt.title("香水价格与评分的关系",fontsize=14)
plt.xlabel("价格（美元）",fontsize=12)
plt.ylabel("评分",fontsize=12)
plt.grid(True,alpha=0.3)
plt.tight_layout()
plt.show()
```
